# Q1. Calibrating Displaced-Diffusion Model

**Calibrate the displaced-diffusion model** to the swaption market data, and document:

### **Q1.0 Helper Functions for Q1 - DD Calibration ; Q2 - SABR Calibration**

*OIS Discount Factor* :   
$$
D_o(t,T) = D_o(t,T_1) + \frac{T - T_1}{T_2 - T_1} \Bigl( D_o(t,T_2) - D_o(t,T_1) \Bigr)
$$

*where*

<ul>
  <li><b>\( D_o(t, T) \)</b> — OIS discount factor from valuation time \( t \) to target maturity \( T \)</li>
  <li><b>\( D_o(t, T_1), D_o(t, T_2) \)</b> — Known OIS discount factors for adjacent maturities \( T_1 \) and \( T_2 \)</li>
  <li><b>\( T \)</b> — The target maturity date for interpolation, such that \( T_1 < T < T_2 \)</li>
  <li><b>\( \frac{T - T_1}{T_2 - T_1} \)</b> — Weighting factor based on the relative time position of \( T \)</li>
</ul>

*PVBP* :
$$
\text{PVBP} = \sum_{i=1}^{n} \Delta_{i-1} \cdot D(t, T_i)
$$

*where:*

<ul>
  <li><b>\( \text{PVBP} \)</b> — Present Value of a Basis Point (per unit notional)</li>
  <li><b>\( \Delta_{i-1} \)</b> — Day count fraction between \( T_{i-1} \) and \( T_i \)</li>
  <li><b>\( D(t, T_i) \)</b> — Discount factor from valuation time \( t \) to payment date \( T_i \)</li>
  <li><b>\( n \)</b> — Number of fixed leg cash flows</li>
</ul>

<ins> **Black76 Formula for Swaptions** </ins>

*Black76 Payer Swaption*
$$
\text{PayerPrice}_{\text{B76}} = \text{PVBP} \cdot \left[ F \cdot \Phi(d_1) - K \cdot \Phi(d_2) \right]
$$


*Black76 Receiver Swaption*
$$
\text{ReceiverPrice}_{\text{B76}} = \text{PVBP} \cdot \left[ K \cdot \Phi(-d_2) - F \cdot \Phi(-d_1) \right]
$$

*where*:

$$
d_1 = \frac{\ln\left( \frac{F}{K} \right) + \frac{1}{2} \sigma^2 T}{\sigma \sqrt{T}}, \quad
d_2 = d_1 - \sigma \sqrt{T}
$$

**Definitions:**

<ul>
  <li><b>\( F \)</b> — Forward swap rate</li>
  <li><b>\( K \)</b> — Strike rate</li>
  <li><b>\( \text{PVBP} \)</b> — Present Value of a Basis Point (annuity factor)</li>
  <li><b>\( \sigma \)</b> — Volatility of forward rate</li>
  <li><b>\( T \)</b> — Time to expiry (in years)</li>
  <li><b>\( \Phi(\cdot) \)</b> — Standard normal cumulative distribution function</li>
</ul>

<ins> **Displaced Diffusion Formula for Swaptions** </ins>

**Adjusted Inputs** :

- Adjusted forward:  
  $$ F^* = \frac{F}{\beta} $$

- Adjusted strike:  
  $$ K^* = K + \left( \frac{1 - \beta}{\beta} \right) F $$

- Adjusted volatility:  
  $$ \sigma^* = \sigma \cdot \beta $$

---

*Displaced Diffusion Payer Swaption*

$$
\text{PayerPrice}_{\text{DD}} = \text{PVBP} \cdot \left[ F^* \cdot \Phi(d_1) - K^* \cdot \Phi(d_2) \right]
$$

*Displaced Diffusion Receiver Swaption*

$$
\text{ReceiverPrice}_{\text{DD}} = \text{PVBP} \cdot \left[ K^* \cdot \Phi(-d_2) - F^* \cdot \Phi(-d_1) \right]
$$

Where:

$$
d_1 = \frac{\ln\left( \frac{F^*}{K^*} \right) + \frac{1}{2} (\sigma^*)^2 T}{\sigma^* \sqrt{T}}, \quad
d_2 = d_1 - \sigma^* \sqrt{T}
$$

**Definitions:**

<ul>
  <li><b>\( F \)</b> — Original forward rate</li>
  <li><b>\( K \)</b> — Strike rate</li>
  <li><b>\( \beta \)</b> — Displacement parameter</li>
  <li><b>\( \sigma \)</b> — Original volatility</li>
  <li><b>\( T \)</b> — Time to maturity (in years)</li>
  <li><b>\( \Phi(\cdot) \)</b> — Standard normal cumulative distribution function</li>
</ul>

<ins>**Root-Finding for Black-76 Implied Volatility using `scipy.optimize.brentq`**</ins>


We define the function:

$$
f(\hat{\sigma}) = V_{\text{B76}}(F, K, \text{PVBP}, \hat{\sigma}, T) - V_{\text{DD}}(F, K, \text{PVBP}, \sigma, T, \beta)
$$

Then the Black-76 implied volatility $ \hat{\sigma} $ is the solution to the root-finding problem:

$$
f(\hat{\sigma}) = 0
$$

Which is equivalent to:

$$
V_{\text{B76}}(F, K, \text{PVBP}, \hat{\sigma}, T) = V_{\text{DD}}(F, K, \text{PVBP}, \sigma, T, \beta)
$$

<ins>**Model Calibration for $ \beta $ using`dd_calibration`**</ins>

We aim to find the displacement parameter $ \beta $ that minimizes the sum of squared errors (SSE) between market and model-implied volatilities:

Where the model-implied volatility is obtained by:

$$
\sigma_{\text{model}}(K_i; \beta) = \text{ImpliedVol}_{\text{B76}} \left( \text{Price}_{\text{DD}}(F, K_i, \text{PVBP}, \sigma_{\text{ATM}}, T, \beta) \right)
$$

**Definitions:**

<ul>
  <li><b>\( \sigma_{\text{market}}(K_i) \)</b> — Market implied vol for strike \( K_i \)</li>
  <li><b>\( \sigma_{\text{ATM}} \)</b> — ATM volatility used for displaced diffusion pricing</li>
  <li><b>\( \text{Price}_{\text{DD}} \)</b> — Option price from displaced diffusion model</li>
  <li><b>\( \text{ImpliedVol}_{\text{B76}} \)</b> — Black-76 implied volatility from option price</li>
  <li><b>\( F \)</b> — Forward rate</li>
  <li><b>\( T \)</b> — Time to maturity</li>
  <li><b>\( \text{PVBP} \)</b> — Annuity factor (present value of a basis point)</li>
</ul>

In words, we do this:

1.  Use **ATM vol** and $ \beta $ to compute the **DD price** at strike $ K_i $

2.  Then invert that price to get a **Black-76 implied vol**

3.  Compare it with the market vol at that strike

4.  Square the error and sum over all strikes


```python
# =============================================================================
# Helper Functions for Q1, Q2.
# =============================================================================

# =============================================================================
# 1) OIS Discount Factor Interpolation and PVBP Calculation. Needed to pass to swaption functions.
# =============================================================================
def ois_discount_interp(tenor):
    """
    Interpolate the OIS discount factor for a given tenor.

    Assumes 'discount_factors_df' (a global DataFrame) has columns:
    'Tenor' and 'OIS Discount Factor'.
    """
    xp = list(df_discount_factors['Tenor'])
    yp = list(df_discount_factors['OIS Discount Factor'])
    return np.interp(tenor, xp, yp)

def calculate_PVBP(expiry, tenor, dcf=0.5):
    """
    Calculate the Present Value of a Basis Point (PVBP) using vectorized operations.

    Parameters:
        expiry (float): Expiry time in years.
        tenor (float): Tenor in years.
        dcf (float): Day count fraction for payment intervals.

    Returns:
        float: PVBP value.
    """
    periods = int(tenor / dcf)
    # Payment times: expiry + dcf, expiry + 2*dcf, ..., expiry + periods*dcf
    payment_times = expiry + dcf * np.arange(1, periods + 1)
    discount_factors = np.interp(payment_times, df_discount_factors['Tenor'], df_discount_factors['OIS Discount Factor'])
    return discount_factors.sum() * dcf

# =============================================================================
# 2) Swaption Pricing Functions For Black76, Displaced Diffusion
# =============================================================================
def black76_payer(F, K, PVBP, sigma, T):
    """
    Price a European call option using the Black-76 formula.

    Parameters:
        F (float): Forward rate.
        K (float): Strike rate.
        PVBP (float): Present Value of a Basis Point (annuity factor).
        sigma (float): Volatility.
        T (float): Time to maturity (in years).

    Returns:
        float: Call option price.
    """
    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return PVBP * (F * norm.cdf(d1) - K * norm.cdf(d2))

def black76_receiver(F, K, PVBP, sigma, T):
    """
    Price a European put option using the Black-76 formula.

    Parameters:
        F (float): Forward rate.
        K (float): Strike rate.
        PVBP (float): Present Value of a Basis Point (annuity factor).
        sigma (float): Volatility.
        T (float): Time to maturity (in years).

    Returns:
        float: Put option price.
    """
    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return PVBP * (K * norm.cdf(-d2) - F * norm.cdf(-d1))

def displaced_diffusion_payer(F, K, PVBP, sigma, T, beta):
    """
    Price a call option under the displaced diffusion model.

    Adjusts the inputs for the Black-76 formula using the displacement parameter.
    """
    adjusted_F = F / beta
    adjusted_K = K + ((1 - beta) / beta) * F
    adjusted_sigma = sigma * beta
    return black76_payer(adjusted_F, adjusted_K, PVBP, adjusted_sigma, T)

def displaced_diffusion_receiver(F, K, PVBP, sigma, T, beta):
    """
    Price a put option under the displaced diffusion model.

    Adjusts the inputs for the Black-76 formula using the displacement parameter.
    """
    adjusted_F = F / beta
    adjusted_K = K + ((1 - beta) / beta) * F
    adjusted_sigma = sigma * beta
    return black76_receiver(adjusted_F, adjusted_K, PVBP, adjusted_sigma, T)

def displaced_diffusion(F, K, PVBP, sigma, T, beta):
    """
    Price an option under the displaced diffusion model.

    Chooses between call and put pricing based on whether the forward rate F
    is less than or equal to the strike K.
    """
    if F <= K:
        # Payer swaption (treated as a call option)
        return displaced_diffusion_payer(F, K, PVBP, sigma, T, beta)
    else:
        # Receiver swaption (treated as a put option)
        return displaced_diffusion_receiver(F, K, PVBP, sigma, T, beta)

# =============================================================================
# 3) Implied Volatility and Calibration Functions
# =============================================================================
def imp_vol(F, K, PVBP, T, sigma, beta):
    """
    Compute the implied volatility under the Black-76 model that matches
    the displaced diffusion model price.

    Parameters:
        F (float): Forward rate.
        K (float): Strike rate.
        PVBP (float): Present Value of a Basis Point.
        T (float): Time to maturity.
        sigma (float): Input volatility (used in displaced diffusion price).
        beta (float): Displacement parameter.

    Returns:
        float: Implied volatility under Black-76.
    """
    if F <= K:
        # For call option pricing
        price = displaced_diffusion_payer(F, K, PVBP, sigma, T, beta)
        iv = brentq(lambda s: price - black76_payer(F, K, PVBP, s, T), 1e-12, 10)
    else:
        # For put option pricing
        price = displaced_diffusion_receiver(F, K, PVBP, sigma, T, beta)
        iv = brentq(lambda s: price - black76_receiver(F, K, PVBP, s, T), 1e-12, 10)
    return iv

def dd_calibration(beta, atm_vol, strikes, market_vols, F, T, PVBP):
    """
    Compute the sum of squared errors between market implied volatilities
    and model-implied volatilities from the displaced diffusion model,
    for a given displacement parameter beta.

    Parameters:
        beta (array-like): Displacement parameter to calibrate.
        atm_vol (float): At-the-money implied volatility.
        strikes (np.array): Array of strike prices.
        market_vols (np.array): Array of market implied volatilities.
        F (float): Forward rate.
        T (float): Time to maturity.
        PVBP (float): Annuity factor.

    Returns:
        float: Sum of squared errors.
    """
    error = 0.0
    for i, market_vol in enumerate(market_vols):
        model_vol = imp_vol(F, strikes[i], PVBP, T, atm_vol, beta)
        error += (market_vol - model_vol) ** 2
    return error

# =============================================================================
# 4) Pivot Table Function to present the calibrated model parameters for given Swaption Expiry and Tenor
# =============================================================================

def create_calibrated_parameter_pivot(dd_results_df, param_name):
  """
  Create a pivot table for a calibrated parameter (e.g., Beta or Sigma)
  by cleaning the 'Tenor' column and pivoting on Expiry and Tenor.

  Parameters:
      dd_results_df (pd.DataFrame): DataFrame containing the calibration results.
                                    Must include columns: "Expiry", "Tenor", and the parameter.
      param_name (str): Name of the parameter column to pivot (e.g., "Beta", "Sigma").

  Returns:
      pd.DataFrame: A pivot table with Expiry as the index and Tenor (with "Y" appended)
                    as columns, and the parameter values as entries.
  """
  # Create a copy of the relevant columns
  df_param = dd_results_df[["Expiry", "Tenor", param_name]].copy()

  # Extract the numeric part of Tenor (e.g., "5Y" -> 5)
  df_param["Tenor"] = df_param["Tenor"].str.extract(r'(\d+)').astype(int)

  # Create a pivot table with Expiry as rows and numeric Tenor as columns
  pivot = df_param.pivot(index="Expiry", columns="Tenor", values=param_name)

  # Reorder rows based on the unique order of Expiry in the original DataFrame
  pivot = pivot.reindex(index=df_param["Expiry"].unique())

  # Rename Tenor columns to include "Y" (e.g., 5 becomes "5Y")
  pivot.columns = [f"{col}Y" for col in pivot.columns]
  pivot.columns.name = "Tenor"

  return pivot

# =============================================================================
# 5) Plot Calibration Results functions.
# =============================================================================

def plot_calibration_results(calib_df, market_df, model_vol_func, model_label,
                            title_prefix="Calibration Plots", extra_calc_func=None,
                            n_cols=5):
    """
    Plot calibration results for a given model.

    Parameters:
    - calib_df: DataFrame with calibration parameters (including Expiry and Tenor).
    - market_df: DataFrame containing market data with columns 'Expiry', 'Tenor', 'strike', 'impliedVol', etc.
    - model_vol_func: Function to compute model volatilities. It should accept (forward, strike, time variables, calibration parameters).
    - model_label: String label for the model curve (e.g. 'DD Model Vols' or 'SABR Vols').
    - title_prefix: Title prefix for the figure's super title.
    - extra_calc_func: Optional function to perform extra computations. This can be used when additional calculations,
                      such as computing PVBP for DD calibration, are necessary. It should return a dict of extra parameters.
    - n_cols: Number of columns in the subplot grid.
    """
    n_plots = len(calib_df)
    n_rows = math.ceil(n_plots / n_cols)

    # Create the subplots grid.
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, 10), tight_layout=True)
    axs_flat = axs.flatten()
    fig.suptitle(title_prefix, fontsize=16)

    # Loop over each calibration result
    for i, (_, row) in enumerate(calib_df.iterrows()):
        # Extract expiry and tenor from the current calibration row.
        expiry = row["Expiry"]
        tenor = row["Tenor"]

        # Filter the market data for the current expiry-tenor combination.
        temp_df = market_df[(market_df["Expiry"] == expiry) & (market_df["Tenor"] == tenor)].copy()
        if temp_df.empty:
            continue  # Skip if no data found

        # Extract common variables.
        temp_F = temp_df["ForwardSwapRate"].iloc[0]
        # Assume "Expiry_no" is used for time; if tenor-specific time is needed, extract that as well.
        temp_T = temp_df["Expiry_no"].iloc[0]

        # Optionally perform extra calculations (e.g., calculate PVBP for displaced diffusion).
        extra_params = {}
        if extra_calc_func is not None:
            extra_params = extra_calc_func(row, temp_df)

        # Compute model-implied volatilities for each strike.
        model_vols = []
        for strike in temp_df['strike']:
            # You can pass extra parameters using keyword arguments if needed.
            vol = model_vol_func(temp_F, strike, temp_T, row, **extra_params)
            model_vols.append(vol)

        # Plotting on the current subplot.
        ax = axs_flat[i]
        ax.plot(temp_df['strike'], temp_df['impliedVol'], 'gs', label='Market Vols')
        ax.plot(temp_df['strike'], model_vols, 'm--', label=model_label)
        ax.set_title(f"{expiry} x {tenor}")
        ax.set_xlabel("Strike")
        ax.set_ylabel("Implied Vol")
        ax.legend()

    # Turn off any unused subplots.
    for j in range(i + 1, len(axs_flat)):
        axs_flat[j].axis('off')

    plt.show()

# =============================================================================
# 5A) Functions used in conjunction with plotting calibration results, specific for Displaced Diffusion
# =============================================================================

# Example usage for Displaced-Diffusion model:
def dd_model_vol_func(F, strike, T, calib_row, **extra_params):
    # Assume extra_params contains 'PVBP'
    PVBP = extra_params.get("PVBP", 1)
    # Call your displaced diffusion volatility function
    return imp_vol(F, strike, PVBP, T, calib_row["Sigma"], calib_row["Beta"])

def dd_extra_calc(calib_row, temp_df):
    # For displaced diffusion, compute PVBP using a separate function.
    temp_expiry_T = temp_df["Expiry_no"].iloc[0]
    temp_tenor_T = temp_df["Tenor_no"].iloc[0]
    PVBP = calculate_PVBP(temp_expiry_T, temp_tenor_T)
    return {"PVBP": PVBP}
```

# Q2. SABR Model Calibration

**Calibrate the SABR model** to the swaption market data using ($\beta = 0.9$), and document:
   - a table of ($\alpha$) parameters  
   - a table of ($\rho$) parameters  
   - a table of ($\nu$) parameters  

### **Q2.0 Helper Functions for Q2 - SABR Calibration**

*We also borrow some helper functions from Q1*


<ins>**SABR Model Implied Volatility (Hagan et al. 2002)**</ins>

<ins>Case 1: ATM (F = K)</ins>

$$
\sigma_{\text{SABR}}(F, F) =
\frac{\alpha}{F^{1 - \beta}} \left[
1 + \left(
\frac{(1 - \beta)^2}{24} \cdot \frac{\alpha^2}{F^{2(1 - \beta)}}
+ \frac{1}{4} \cdot \frac{\rho \beta \nu \alpha}{F^{1 - \beta}}
+ \frac{2 - 3\rho^2}{24} \cdot \nu^2
\right) T
\right]
$$

---

<ins>Case 2: Non-ATM (F ≠ K)</ins>

Define:

$$
z = \frac{\nu}{\alpha} \cdot (F K)^{\frac{1 - \beta}{2}} \cdot \ln\left(\frac{F}{K}\right)
$$

$$
\chi(z) = \ln\left( \frac{\sqrt{1 - 2\rho z + z^2} + z - \rho}{1 - \rho} \right)
$$

Then the SABR implied volatility is:

$$
\sigma_{\text{SABR}}(F, K) =
\frac{
\alpha \left[ 1 + \left( A_1 + A_2 + A_3 \right) T \right] \cdot z
}{
(F K)^{\frac{1 - \beta}{2}} \cdot \left[ 1 + B_1 + B_2 \right] \cdot \chi(z)
}
$$

Where:

- $$A_1 = \frac{(1 - \beta)^2}{24} \cdot \frac{\alpha^2}{(F K)^{1 - \beta}}$$
- $$A_2 = \frac{1}{4} \cdot \frac{\rho \beta \nu \alpha}{(F K)^{(1 - \beta)/2}}$$
- $$A_3 = \frac{2 - 3\rho^2}{24} \cdot \nu^2$$

- $$B_1 = \frac{(1 - \beta)^2}{24} \cdot \left[\ln\left(\frac{F}{K}\right)\right]^2$$
- $$B_2 = \frac{(1 - \beta)^4}{1920} \cdot \left[\ln\left(\frac{F}{K}\right)\right]^4$$


### SABR Calibration Objective

We calibrate the SABR model parameters $ \alpha $, $ \rho $, and $ \nu $ (with $ \beta $ fixed) by minimizing the sum of squared errors:

$$
\text{SSE}(\alpha, \rho, \nu) = \sum_{i=1}^{n} \left[
\sigma_{\text{market}}(K_i) - \sigma_{\text{SABR}}(K_i; \alpha, \rho, \nu, \beta)
\right]^2
$$

**Where:**

<ul>
  <li><b>$ \sigma_{\text{market}}(K_i) $</b> — Market implied volatility at strike $ K_i $</li>
  <li><b>$ \sigma_{\text{SABR}}(K_i; \alpha, \rho, \nu, \beta) $</b> — SABR model-implied volatility at strike $ K_i $</li>
  <li><b>$ \alpha $</b> — Initial volatility (SABR alpha)</li>
  <li><b>$ \rho $</b> — Correlation between forward and volatility (SABR rho)</li>
  <li><b>$ \nu $</b> — Volatility of volatility (SABR nu)</li>
  <li><b>$ \beta $</b> — Elasticity parameter (assumed fixed)</li>
  <li><b>$ F $</b> — Forward rate</li>
  <li><b>$ T $</b> — Time to maturity</li>
</ul>



```python
# =============================================================================
# 1) SABR Model Functions
# =============================================================================
def sabr_implied_vol(F, K, T, alpha, beta, rho, nu):
    """
    Compute the SABR model implied volatility given market inputs.

    Parameters:
        F (float): Forward rate.
        K (float): Strike rate.
        T (float): Time to maturity (years).
        alpha (float): SABR alpha parameter.
        beta (float): SABR beta parameter.
        rho (float): SABR rho parameter.
        nu (float): SABR nu parameter (volatility of volatility).

    Returns:
        float: Implied volatility computed by the SABR model.
    """
    # Use K directly as X
    X = K

    # At-the-money (ATM) case
    if abs(F - K) < 1e-12:
        numer1 = (((1 - beta) ** 2) / 24) * (alpha ** 2) / (F ** (2 - 2 * beta))
        numer2 = 0.25 * rho * beta * nu * alpha / (F ** (1 - beta))
        numer3 = ((2 - 3 * rho ** 2) / 24) * (nu ** 2)
        VolAtm = alpha * (1 + (numer1 + numer2 + numer3) * T) / (F ** (1 - beta))
        sabr_sigma = VolAtm
    else:
        # Non-ATM case: compute the intermediate variables
        z = (nu / alpha) * ((F * X) ** (0.5 * (1 - beta))) * np.log(F / X)
        zhi = np.log((np.sqrt(1 - 2 * rho * z + z ** 2) + z - rho) / (1 - rho))
        numer1 = (((1 - beta) ** 2) / 24) * (alpha ** 2) / ((F * X) ** (1 - beta))
        numer2 = 0.25 * rho * beta * nu * alpha / ((F * X) ** ((1 - beta) / 2))
        numer3 = ((2 - 3 * rho ** 2) / 24) * (nu ** 2)
        numer = alpha * (1 + (numer1 + numer2 + numer3) * T) * z

        denom1 = ((1 - beta) ** 2 / 24) * (np.log(F / X)) ** 2
        denom2 = (((1 - beta) ** 4) / 1920) * (np.log(F / X)) ** 4
        denom = ((F * X) ** ((1 - beta) / 2)) * (1 + denom1 + denom2) * zhi

        sabr_sigma = numer / denom

    return sabr_sigma


def sabr_calibration_error(params, strikes, market_vols, F, T):
    """
    Computes the sum of squared errors between the SABR model volatilities and
    market volatilities for a given set of parameters.

    Parameters:
        params (array-like): Array containing [alpha, rho, nu].
        strikes (np.array): Array of strike prices.
        market_vols (np.array): Array of market implied volatilities.
        F (float): Forward rate.
        T (float): Time to maturity.

    Returns:
        float: Sum of squared errors.
    """
    error = 0.0
    alpha, rho, nu = params
    for i, market_vol in enumerate(market_vols):
        model_vol = sabr_implied_vol(F, strikes[i], T, alpha, float_given_beta, rho, nu)
        error += (market_vol - model_vol) ** 2
    return error


# =============================================================================
# 5A) Functions used in conjunction with plotting calibration results, specific for Displaced Diffusion
# =============================================================================

def sabr_model_vol_func(F, strike, T, calib_row, **extra_params):
    # In this case, no extra parameters are needed.
    # Assume 'float_given_beta' is defined globally or available in the scope.
    return sabr_implied_vol(F, strike, T, calib_row["Alpha"], float_given_beta, calib_row["Rho"], calib_row["Nu"])

```

```python
# =============================================================================
# SABR Calibration Loop
# =============================================================================
sabr_results = []
initial_guess = [0.02, 0.2, 0.1]  # Initial Guesses for [alpha, rho, nu] for brentq
float_given_beta = 0.9 # Global constant for SABR beta (as per project instructions )

# Loop over each expiry and tenor combination (assumed array_expiry and array_tenor are defined)
for expiry in array_expiry:
    for tenor in array_tenor:
        # Filter the data for the current expiry-tenor pair
        df_calibrate = df_swaption_fwd_swap_rate[(df_swaption_fwd_swap_rate['Expiry'] == expiry) & (df_swaption_fwd_swap_rate['Tenor'] == tenor)].copy()

        # Extract time-to-maturity (T) and forward par swap rate (F)
        T = df_calibrate['Expiry_no'].iloc[0]
        F = df_calibrate['ForwardSwapRate'].iloc[0]

        # Calibrate SABR parameters by minimizing the calibration error
        res = least_squares(
            lambda params: sabr_calibration_error(params,
                                                  df_calibrate['strike'].values,
                                                  df_calibrate['impliedVol'].values,
                                                  F, T),
            initial_guess,
            bounds=([0, -1, 0], [np.inf, 1, np.inf])
        )
        alpha_cal = res.x[0]
        rho_cal = res.x[1]
        nu_cal = res.x[2]

        # Append calibration result for this expiry-tenor pair
        sabr_results.append([expiry, tenor, alpha_cal, rho_cal, nu_cal])
```

# Q3. Swaption Pricing

**Price the following swaptions** using the calibrated displaced-diffusion and SABR model:

### **Q3.0 Helper Functions for Q3**

*Forward LIBOR Rate* :
$$
L_{t}(T_1,T_2) = \frac{1}{\triangle(T_1,T_2)}\left(\frac{D(t,T_1)}{D(t,T_2)} - 1\right)
$$

where: $$
t < T_1 < T_2
$$

*Linear Interpolation of LIBOR Disount Factor* :
$$
D(t,T) = D(t,T_1) + \frac{T - T_1}{T_2 - T_1} \Bigl( D(t,T_2) - D(t,T_1) \Bigr)
$$

*PV of floating payments of a swap* :
$$
S_{n,N}(0) = \frac{\text{PV}_{\text{flt}}}{\text{PVBP}}
= \frac{D(0, T_0) - D(0, T_n)}{\sum_{i=1}^{n} \Delta_{i-1} D(0, T_i)}
$$

Where:

<ul>
  <li><b>\( \Delta_i \)</b> — Day count fraction between \( T_{i-1} \) and \( T_i \)</li>
  <li><b>\( L(T_{i-1}, T_i) \)</b> — Forward LIBOR rate from \( T_{i-1} \) to \( T_i \)</li>
  <li><b>\( D(t, T_i) \)</b> — Discount factor from time \( t \) to payment date \( T_i \)</li>
  <li><b>\( n \)</b> — Number of floating cash flows</li>
</ul>

*Par Swap Rate* :
$$
S_{n,N}(0) = \frac{\text{PV}_{\text{flt}}}{\text{PVBP}}
= \frac{D(0, T_0) - D(0, T_n)}{\sum_{i=1}^{n} \Delta_{i-1} D(0, T_i)}
$$

Where:

<ul>
  <li><b>\( S_{n,N}(0) \)</b> — Par swap rate at time 0 for a swap from \( T_0 \) to \( T_n \)</li>
  <li><b>\( D(0, T_i) \)</b> — Discount factor from time 0 to payment date \( T_i \)</li>
  <li><b>\( \Delta_{i-1} \)</b> — Day count fraction between \( T_{i-1} \) and \( T_i \)</li>
  <li><b>\( \text{PVBP} \)</b> — Present Value of a Basis Point: \( \sum_{i=1}^{n} \Delta_{i-1} D(0, T_i) \)</li>
  <li><b>\( \text{PV}_{\text{flt}} \)</b> — Present value of the floating leg: \( D(0, T_0) - D(0, T_n) \)</li>
  <li><b>\( D(0, T_0) = 1 \)</b>
</ul>

**`interpolate_model_parameters` Interpolation Function**

<ins>Explanation of `interpolate_model_parameters`</ins>

The `interpolate_model_parameters` function interpolates for model parameters at a target (non-market-observable) expiry **T** and tenor **N**. The input data is expected to be a pivoted DataFrame where:

- **Index:** Contains expiry values (e.g., "1Y", "2Y", etc.).
- **Columns:** Contain tenor values (e.g., "1Y", "2Y", etc.).
- **Data:** Contains the calibrated parameter values corresponding to each expiry-tenor combination.

---

<ins>Interpolation Methods</ins>

The function supports different interpolation methods. Two main approaches are:

<ins>A. Cubic Spline Interpolation (`method == 'cubic'`)</ins>

  *Interpolates along the expiry axis for each tenor, then across tenors.*

1. **Interpolate Along the Expiry Dimension:**
   - For each tenor column, a cubic spline is created using the numeric expiry values as the independent variable and the corresponding parameter values as the dependent variable.
   - The spline is evaluated at the target expiry **T** to obtain an interpolated value for that tenor.
   - The results for all tenor columns are stored in a list.

2. **Interpolate Along the Tenor Dimension:**
   - The function pairs each tenor (numeric) with its corresponding interpolated value from the previous step.
   - These pairs are sorted by tenor.
   - A second cubic spline is constructed using the sorted tenor values and the interpolated parameter values.
   - This spline is then evaluated at the target tenor **N**, yielding the final interpolated parameter value.

<ins>B. Regular Grid Interpolation (using methods like `"linear"`, `"nearest"`, `"slinear"`, or `"pchip"`)</ins>

*Directly interpolates on a sorted 2D grid using methods such as linear or nearest neighbor interpolation.*

1. **Prepare a 2D Grid:**
   - The parameter values are arranged into a 2D array (grid) from the DataFrame.
   - Both the expiry and tenor arrays are sorted.
   - The parameter grid is re-ordered to match the sorted expiry and tenor values.

2. **Interpolate on the 2D Grid:**
   - The function sets up a `RegularGridInterpolator` using the sorted expiry and tenor arrays as grid axes.
   - The sorted parameter grid is passed as data.
   - The chosen interpolation method is applied.
   - The interpolator estimates the parameter value at the point \((T, N)\).

- **Output:**  interpolated parameter value at the specified expiry **T** and tenor **N**.


```python
# =============================================================================
# 1) LIBOR related Formula

# LIBOR Forward Rate ;
# LIBOR Discount Factor Interpolation (by linear interpolation)
# =============================================================================
def libor_forward(D1, D2, delta=0.5):
    """
    Calculate the LIBOR forward rate from two discount factors.
    """
    return (D1 / D2 - 1) / delta

def libor_discount_interp(tenor):
    """
    Interpolate the LIBOR Discount Factors for a given tenor.
    """
    xp = df_discount_factors['Tenor']
    yp = df_discount_factors['LIBOR Discount Factor']
    return np.interp(tenor, xp, yp)

# =============================================================================
# 2) Swap Related Formulas: PV_flt & Par Swap Rate
# =============================================================================

def calculate_pv_flt(expiry, tenor, delta=0.5):
    """
    Calculate the pv of floating leg of a swap using vectorized operations.

    This computes the sum over each semiannual payment period from expiry to expiry+tenor,
    using the product of the OIS discount factor and the LIBOR forward rate.

    Parameters:
        expiry (float): Start time (years) for the floating leg.
        tenor (float): Length (years) of the swap.
        delta (float): Payment interval (day count fraction) (years), default is 0.5 for semiannual payments.

    Returns:
        float: The present value of the floating leg.
    """
    # Number of payment periods
    m = int(tenor / delta)

    # Create an array for payment period indices: 1, 2, ..., m
    periods = np.arange(1, m + 1)

    # Compute payment times: expiry + i * delta
    payment_times = expiry + periods * delta

    # Interpolate OIS discount factors for each payment time
    disc_factors = np.interp(payment_times,
                             df_discount_factors['Tenor'],
                             df_discount_factors['OIS Discount Factor'])

    # Interpolate LIBOR discount factors for payment times and one period earlier
    D1 = np.interp(payment_times - delta,
                   df_discount_factors['Tenor'],
                   df_discount_factors['LIBOR Discount Factor'])
    D2 = np.interp(payment_times,
                   df_discount_factors['Tenor'],
                   df_discount_factors['LIBOR Discount Factor'])

    # Calculate forward LIBOR rates for each period: (D1/D2 - 1)/delta
    forward_rates = (D1 / D2 - 1) / delta

    # Compute the sum of the discounted forward rates and multiply by delta
    return np.sum(disc_factors * forward_rates) * delta


def swaption_rate(expiry, tenor):
    """
    Calculate the par swap rate for a given expiry and tenor.

    It is the ratio of the float leg value to the PVBP.
    """
    return calculate_pv_flt(expiry, tenor) / calculate_PVBP(expiry, tenor)

# =============================================================================
# 3) Model Calibrated Parameters Interpolation.

"""Allows choice of use of cubic spline, linear, nearest, slinear interpolation"""
# =============================================================================


def interpolate_model_parameters(T, N, data, method='slinear'):
    """
    Interpolate calibrated SABR / DD parameters using either cubic spline interpolation
    or linear / nearest / slinear interpolation via RegularGridInterpolator.

    Parameters:
        T (float): Target expiry (years).
        N (float): Target tenor (years).
        data (pd.DataFrame): Pivoted DataFrame of calibrated parameters. The index and columns
                             may be strings (e.g., "1Y") or numeric.
        method (str): Interpolation method. Options: ["linear", "nearest", "slinear", "cubic", "quintic" and "pchip"]

    Returns:
        float: Interpolated parameter value at expiry T and tenor N.
    """
    # Extract numeric expiry values from the index
    expiry_vals = []
    for idx in data.index.tolist():
        if isinstance(idx, str):
            numeric_part = ''.join(filter(str.isdigit, idx))
            expiry_vals.append(int(numeric_part))
        else:
            expiry_vals.append(int(idx))
    expiry_vals = np.array(expiry_vals)

    # Extract numeric tenor values from the columns
    tenor_vals = []
    for col in data.columns:
        if isinstance(col, str):
            numeric_part = ''.join(filter(str.isdigit, col))
            tenor_vals.append(int(numeric_part))
        else:
            tenor_vals.append(int(col))
    tenor_vals = np.array(tenor_vals)

    if method == 'cubic':
        # For each tenor column, interpolate along expiry dimension using CubicSpline.
        interpolated_expiry = []
        for col in data.columns:
            cs = CubicSpline(expiry_vals, data[col], bc_type=((2, 0.0), (2, 0.0)), extrapolate=True)
            interpolated_expiry.append(cs(T))

        # Ensure sorting by tenor values
        sorted_pairs = sorted(zip(tenor_vals, interpolated_expiry), key=lambda x: x[0])
        tenor_sorted, interp_sorted = zip(*sorted_pairs)

        # Interpolate along tenor dimension
        cs_tenor = CubicSpline(tenor_sorted, interp_sorted, bc_type=((2, 0.0), (2, 0.0)), extrapolate=True)
        return float(cs_tenor(N))

    elif method in ["linear","nearest", "slinear", "pchip"]:
        # Prepare a 2D grid of parameter values.
        parameter_grid = data.values.astype(float)
        # Ensure grid is sorted by expiry and tenor
        # Sort expiry and tenor if not already sorted.
        expiry_sort_idx = np.argsort(expiry_vals)
        tenor_sort_idx = np.argsort(tenor_vals)
        expiry_sorted = np.sort(expiry_vals)
        tenor_sorted = np.sort(tenor_vals)
        parameter_grid_sorted = parameter_grid[np.ix_(expiry_sort_idx, tenor_sort_idx)]

        interpolator = RegularGridInterpolator((expiry_sorted, tenor_sorted),
                                               parameter_grid_sorted,
                                               method=method,
                                               bounds_error=False,
                                               fill_value=None)
        return float(interpolator([[T, N]])[0])
    else:
        raise ValueError("Invalid method. Choose 'linear', 'nearest', 'slinear'.")

# =============================================================================
# 3) Swaption Prices depending on Expiry and Teno, and Option type i.e. Receiver or Payer

"""User can also specify method for interpolation method"""
# =============================================================================

def compute_swaption_prices(expiry, tenor, option_type, method='slinear'):
    """
    Compute swaption prices (both displaced-diffusion and SABR) for a given expiry and tenor.

    Parameters:
        expiry (float): Expiry of the swaption in years.
        tenor (float): Underlying swap tenor in years.
        option_type (str): Either "Payer" or "Receiver".
        method (str): Interpolation method for model parameters (e.g., 'slinear', 'cubic', 'linear', etc.).

    Returns:
        tuple: Two dictionaries corresponding to the two rows for the output DataFrame:
               one for the displaced diffusion (DD) price and one for the SABR price.
               Each dictionary has an "OptionType" key and keys for each strike in array_K.
    """
    # Get the calibrated model parameters via interpolation using the specified method.
    float_alpha = interpolate_model_parameters(expiry, tenor, df_alpha_pivot, method=method)
    float_rho   = interpolate_model_parameters(expiry, tenor, df_rho_pivot, method=method)
    float_nu    = interpolate_model_parameters(expiry, tenor, df_nu_pivot, method=method)
    float_beta  = interpolate_model_parameters(expiry, tenor, df_beta_pivot, method=method)
    float_sigma = interpolate_model_parameters(expiry, tenor, df_sigma_pivot, method=method)

    # Compute the forward (par swap) rate for the swaption.
    forward_rate = swaption_rate(expiry, tenor)

    prices_dd = []
    prices_sabr = []

    # Loop over each strike in array_K.
    for k in array_K:
        PVBP_val = calculate_PVBP(expiry, tenor)
        if option_type.lower() == 'payer':
            # For payer swaption, use the payer pricing functions.
            price_dd   = displaced_diffusion_payer(forward_rate, k, PVBP_val, float_sigma, expiry, float_beta)
            sabr_vol   = sabr_implied_vol(forward_rate, k, expiry, float_alpha, float_given_beta, float_rho, float_nu)
            price_sabr = black76_payer(forward_rate, k, PVBP_val, sabr_vol, expiry)
        elif option_type.lower() == 'receiver':
            # For receiver swaption, use the receiver pricing functions.
            price_dd   = displaced_diffusion_receiver(forward_rate, k, PVBP_val, float_sigma, expiry, float_beta)
            sabr_vol   = sabr_implied_vol(forward_rate, k, expiry, float_alpha, float_given_beta, float_rho, float_nu)
            price_sabr = black76_receiver(forward_rate, k, PVBP_val, sabr_vol, expiry)
        else:
            raise ValueError("option_type must be either 'Payer' or 'Receiver'")

        prices_dd.append(price_dd)
        prices_sabr.append(price_sabr)

    # Build dictionary rows with an "OptionType" field.
    row_dd = {"OptionType": f"{option_type} ({expiry}y x {tenor}y) DD"}
    row_sabr = {"OptionType": f"{option_type} ({expiry}y x {tenor}y) SABR"}
    for idx, k in enumerate(array_K):
        row_dd[k] = prices_dd[idx]
        row_sabr[k] = prices_sabr[idx]

    return row_dd, row_sabr

# =============================================================================
# 4) Present Swaption Prices Results in Pivoted Dataframe
# =============================================================================
def build_swaption_prices_df(method='slinear'):
    """
    Build a swaption prices DataFrame by computing prices for both payer and receiver swaptions.

    The function computes:
      - Payer swaption (2y x 10y)
      - Receiver swaption (8y x 10y)

    Parameters:
        method (str): The interpolation method to use when fetching model parameters.

    Returns:
        pd.DataFrame: DataFrame with an "OptionType" column and one column per strike level.
    """
    rows = []

    # Compute prices for Payer (2y x 10y)
    row_payer_dd, row_payer_sabr = compute_swaption_prices(2, 10, "Payer", method=method)
    # Compute prices for Receiver (8y x 10y)
    row_receiver_dd, row_receiver_sabr = compute_swaption_prices(8, 10, "Receiver", method=method)

    # Combine all rows.
    rows.extend([row_payer_dd, row_payer_sabr, row_receiver_dd, row_receiver_sabr])

    # Create a DataFrame from the rows.
    df = pd.DataFrame(rows)
    return df

# =============================================================================
# 5) Plotting function for swaption prices
# =============================================================================

def plot_swaption_prices_ax(df, ax):
    """
    Plot swaption prices from a DataFrame on a provided Axes,
    and annotate each data point with its (strike, price) coordinates.

    Parameters:
        df (pd.DataFrame): Swaption prices DataFrame. Must have an 'OptionType' column
                           and additional columns for strike levels.
        ax (matplotlib.axes.Axes): The Axes object to plot on.
    """
    # Identify the strike columns (all columns except 'OptionType')
    strike_cols = [col for col in df.columns if col != 'OptionType']

    # Determine if the strike columns are numeric or string.
    # Convert to float and sort.
    if isinstance(strike_cols[0], str):
        strikes = sorted([float(k) for k in strike_cols])
        # Use string keys to extract values from the DataFrame
        col_keys = [str(k) for k in strikes]
    else:
        strikes = sorted([float(k) for k in strike_cols])
        col_keys = strikes

    # Plot each row (each option type) on the given axes.
    for _, row in df.iterrows():
        option_type = row['OptionType']
        # Extract the prices corresponding to sorted strikes.
        prices = [row[k] for k in col_keys]
        ax.plot(strikes, prices, marker='o', label=option_type)

        # Annotate each point with its coordinates
        for x, y in zip(strikes, prices):
            # Format the annotation text with strike and price rounded to 4 decimals.
            ax.annotate(f"({x:.2f}, {y:.4f})", (x, y),
                        textcoords="offset points", xytext=(0, 5),
                        ha='center', fontsize=8)

    ax.set_xlabel('Strike')
    ax.set_ylabel('Swaption Price')
    ax.grid(True)
    ax.legend(title='Option Type', fontsize='small')
```
